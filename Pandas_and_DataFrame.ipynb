```python
# for demonstrating OOP
class Point:
    x=0;
    y=0;
    def __init__(self,x,y):
        self.x=x;
        self.y=y;
        
    def show(self):
        print(self.x,",",self.y)
        
v = Point(3,4) #constructor
v.show()
```

```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
# list of courses
lst = ['Computer Science', 'Commerce PA', 'English', 'Mathematics', 'Microbiology', 'Biochemistry']
print(type(lst))
# Calling DataFrame constructor on list
df = pd.DataFrame(lst)
print(type(df))
print(df[0][0])
print(df)
```

```python
# intialise data of lists for dataframe
data = {'Course':['Computer Science', 'Commerce PA', 'English', 'Mathematics', 'Microbiology', 'Biochemistry'],
        'Strength':[210, 221, 134, 185, 123, 124],
        'Block':['B','A','A','B','A','A']}
type(data)
data
```

```python
# Create DataFrame
df = pd.DataFrame(data)
# Print the output.
print(df)
```
